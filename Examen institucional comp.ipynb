```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Función para mostrar el menú interactivo
def display_menu():
    clear_output()  # Limpia pantalla
    print("Seleccionar una opción:")

    # Crear botones para cada opción
    button_datos = widgets.Button(description="Visualizar datos")
    button_VPN_TIR = widgets.Button(description="Calcular VPN/TIR")
    button_resultados = widgets.Button(description="Graficar resultados")
    button_analisis = widgets.Button(description="Analisis")
    button_salir = widgets.Button(description="Salir")

    # Funciones que se ejecutan cuando se hace clic en cada botón
    def on_button_datos_clicked(b):
        print("Visualizar datos seleccionados.")

    def on_button_VPN_TIR_clicked(b):
        print("Calcular VPN/TIR seleccionado.")

    def on_button_resultados_clicked(b):
        print("Graficar resultados seleccionados.")

    def on_button_analisis_clicked(b):
        print("Analisis resultados seleccionado.")

    def on_button_salir_clicked(b):
        clear_output()
        print("Cerrado el programa...")

    # Conectar cada botón con su respectiva función
    button_datos.on_click(on_button_datos_clicked)
    button_VPN_TIR.on_click(on_button_VPN_TIR_clicked)
    button_resultados.on_click(on_button_resultados_clicked)
    button_analisis.on_click(on_button_analisis_clicked)
    button_salir.on_click(on_button_salir_clicked)

    # Mostrar los botones
    display(button_datos, button_VPN_TIR, button_resultados,button_analisis, button_salir)

# Ejecutar el menú
display_menu()
```

```python
def calcular_vpn(df, tasa_descuento):
    columnas_flujo = [col for col in df.columns if col.startswith('Inversion') or col.startswith('Ano')]
    vpn = []
    for proyecto in df.index:
        flujos = df.loc[proyecto, columnas_flujo].astype(float).values
        valor_vpn = npf.npv(tasa_descuento, flujos)
        vpn.append(round(valor_vpn, 2))
    df['VPN'] = vpn
    return df

def calcular_tir(df):
    columnas_flujo = [col for col in df.columns if col.startswith('Inversion') or col.startswith('Ano')]
    tir = []
    for proyecto in df.index:
        flujos = df.loc[proyecto, columnas_flujo].astype(float).values
        valor_tir = npf.irr(flujos)
        tir.append(round(valor_tir, 4))
    df['TIR'] = tir
    return df
```

```python
import matplotlib.pyplot as plt

def graficar_resultados_matplotlib(df, tasa_descuento=0.1):
    proyectos = df.index.tolist()
    vpn = df['VPN'].tolist()
    tir = df['TIR'].tolist()

    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    # Gráfico de VPN
    axs[0].bar(proyectos, vpn, color='skyblue')
    axs[0].axhline(0, color='red', linestyle='--')
    axs[0].set_title('Valor Presente Neto (VPN) por Proyecto')
    axs[0].set_xlabel('Proyecto')
    axs[0].set_ylabel('VPN ($)')
    axs[0].tick_params(axis='x', rotation=45)
    for i, valor in enumerate(vpn):
        axs[0].text(i, valor + (0.05 * max(vpn)), f'{valor:.0f}', ha='center', va='bottom', fontsize=8)

    # Gráfico de TIR
    axs[1].bar(proyectos, tir, color='lightgreen')
    axs[1].axhline(tasa_descuento, color='orange', linestyle='--', label=f'Tasa de Descuento ({tasa_descuento*100:.0f}%)')
    axs[1].set_title('Tasa Interna de Retorno (TIR) por Proyecto')
    axs[1].set_xlabel('Proyecto')
    axs[1].set_ylabel('TIR (%)')
    axs[1].tick_params(axis='x', rotation=45)
    axs[1].legend()
    for i, valor in enumerate(tir):
        axs[1].text(i, valor + (0.01 * max(tir)), f'{valor:.2%}', ha='center', va='bottom', fontsize=8)

    plt.tight_layout()
    plt.show()
```

```python
def menu_interactivo():
    while True:
        print("\n--- Menú de Análisis de Proyectos ---")
        print("1 Visualizar Datos")
        print("2 Calcular VPN y TIR")
        print("3 Graficar Resultados")
        print("4 Salir")

        opcion = input("Selecciona una opción (1-4): ")

        if opcion == '1':
            display(df.head())
        elif opcion == '2':
            tasa_descuento = 0.1
            calcular_vpn(df, tasa_descuento)
            calcular_tir(df)
            print("Cálculo completado. Resultados:")
            display(df[['VPN', 'TIR']])
        elif opcion == '3':
            graficar_resultados(df)
        elif opcion == '4':
            print("Saliendo del menú.")
            break
        else:
            print("Opción no válida. Intentá de nuevo.")
```
